## Binary Classification
#### Who is likely to have ever used marijuana?

```python
#Divided imp. features into demographic,school and parental 
import numpy as np
demographic_features = ['IRSEX', 'NEWRACE2', 'INCOME', 'POVERTY3', 'PDEN10', 'COUTYP4', 'HEALTH2']
school_features = ['EDUSCHLGO', 'EDUSKPCOM', 'EDUSCHGRD2']
parental_features = ['IMOTHER', 'IFATHER', 'PARCHKHW', 'PARHLPHW', 'PRCHORE2', 'PRLMTTV2', 'PARLMTSN',
                     'PRGDJOB2', 'PRPROUD2', 'ARGUPAR']


all_features = demographic_features + school_features + parental_features

#replaced 991 or 993 (refer codebook) with NA and droppped NA
X_1 = df[all_features].replace([991, 993], np.nan).dropna()
y_1 = df.loc[X_1.index, 'MRJFLAG']
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree, DecisionTreeRegressor
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
from sklearn.model_selection import train_test_split, GridSearchCV
tree_drug = DecisionTreeClassifier()
tree_drug.fit(X_1, y_1)
```

```python
tree_summary = export_text(tree_drug, feature_names=X_1.columns.tolist())
print(tree_summary)
```

```python
#Importances of given predictors in the above tree
importances = pd.DataFrame({'feature_name': readable_features, 'importance': tree_drug.feature_importances_})
importances = importances.sort_values('importance', ascending=False).reset_index(drop=True)
print(importances)
```

```python
feature_names_map = {
    'IRSEX': 'Gender',
    'NEWRACE2': 'Race/Ethnicity',
    'INCOME': 'Family Income',
    'POVERTY3': 'Poverty Level',
    'PDEN10': 'Population Density',
    'COUTYP4': 'County Type',
    'HEALTH2': 'Health Rating',
    'EDUSCHLGO': 'Currently in School',
    'EDUSCHGRD2': 'School Grade',
    'EDUSKPCOM': 'Missed School Days (Sick)',
    'IMOTHER': 'Mother in Household',
    'IFATHER': 'Father in Household',
    'PARCHKHW': 'Parent Checks Homework',
    'PARHLPHW': 'Parent Helps Homework',
    'PRCHORE2': 'Parent Assigns Chores',
    'PRLMTTV2': 'Limits on TV Time',
    'PARLMTSN': 'Limits on Going Out',
    'PRGDJOB2': 'Parental Praise (Job)',
    'PRPROUD2': 'Parental Pride',
    'ARGUPAR': 'Family Arguments',
}

readable_features = [feature_names_map.get(col, col) for col in X_1.columns]


plt.figure(figsize=(50,50))
plot_tree(tree_drug
          , filled=True
          , feature_names=readable_features
          , class_names=["Never Used", "Used"]
          , label='all'
          , fontsize=12)
plt.show()
```

```python
# Split the Data between test and train
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1
                                                    , train_size = 0.5, test_size = 0.5
                                                    , random_state = 2)
# predict on test data
tree_pred = tree_drug.predict(X_test)

# create confusion matrix
confusion_matrix = pd.crosstab(index=tree_pred, columns=y_test, rownames=[''])
print(confusion_matrix)

accuracy = tree_drug.score(X_test, y_test)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
# cross-validation to determine optimal tree size
params = {'max_leaf_nodes': range(2, 20)}  
cv_drug = GridSearchCV(tree_drug, params, cv=10)
cv_drug.fit(X_train, y_train)
cv_results = cv_drug.cv_results_

# find the best score for max leaf nodes
best_size = cv_drug.best_params_['max_leaf_nodes']
best_score = cv_drug.best_score_

# plot results of cross-validation
plt.figure(figsize=(6, 4))
plt.plot(cv_results["param_max_leaf_nodes"], cv_results["mean_test_score"], 'o-')
plt.plot(best_size, best_score, 'ro-')
plt.xlabel('Tree Size')
plt.ylabel('Accuracy')
plt.title('Cross-validation Results')
```

#### predict marijuana use (yes/no) from all three categories of factors â€” (binary classification model) and measuring Feature Importance

```python
demographic_features = ['IRSEX', 'NEWRACE2', 'INCOME', 'POVERTY3', 'PDEN10', 'COUTYP4', 'HEALTH2']
school_features = ['EDUSCHLGO', 'EDUSCHGRD2', 'EDUSKPCOM']
parental_features = ['IMOTHER', 'IFATHER', 'PARCHKHW', 'PARHLPHW', 'PRCHORE2', 'PRLMTTV2', 'PARLMTSN',
                     'PRGDJOB2', 'PRPROUD2', 'ARGUPAR']

all_features = demographic_features + school_features + parental_features

from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = df[all_features].replace([991, 993], np.nan).dropna()
y = df.loc[X.index, 'MRJFLAG']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))

import matplotlib.pyplot as plt
import seaborn as sns

importances = pd.Series(model.feature_importances_, index=readable_features).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances.values, y=importances.index)
plt.title('Feature Importance for Predicting Marijuana Use (MRJFLAG)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
def group_importance(importance_series, feature_groups):
    group_scores = {}
    for group, features in feature_groups.items():
        group_scores[group] = importance_series[importance_series.index.isin(features)].sum()
    return pd.Series(group_scores).sort_values(ascending=False)

feature_groups = {
    'Demographic': demographic_features,
    'School': school_features,
    'Parental': parental_features
}

group_scores = group_importance(importances, feature_groups)
print(group_scores)
```

#### Using RandomForest we get 84% which is approx. same as optimal tree size
Based on group imp. for RandomForest we can see that Demographic features plays a major role in predicting use of marijuana followed by Parental and School.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Your features
all_features = demographic_features + school_features + parental_features

# Clean data
df_clean = df.copy()
df_clean[all_features] = df_clean[all_features].replace([991, 993], np.nan)
df_clean = df_clean.dropna(subset=all_features + ['MRJYDAYS'])

# Treat MRJYDAYS as a category
X = df_clean[all_features]
y = df_clean['MRJYDAYS'].astype(int)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Fit multi-class classifier
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluate
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

importances = pd.DataFrame({'feature_name': readable_features, 'importance': model.feature_importances_})
importances = importances.sort_values('importance', ascending=False).reset_index(drop=True)
print(importances)

importances = pd.Series(model.feature_importances_, index=readable_features).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances.values, y=importances.index)
plt.title('Feature Importance for Predicting Marijuana Use (MRJFLAG)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

The above classfication report is a classic example of class imbalance which is clearly seen as class 6 has 93% accuracy 
and the remaining classes has 0% accuracy.
To overcome this issue I have used SMOTE. This technique increases new synthetic samples in minority class.

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train, y_train)

model = RandomForestClassifier(random_state=42)
model.fit(X_resampled, y_resampled)
```

```python
y_pred = model.predict(X_test)

# Evaluate
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

importances = pd.Series(model.feature_importances_, index=readable_features).sort_values(ascending=False)



plt.figure(figsize=(10, 6))
sns.barplot(x=importances.values, y=importances.index)
plt.title('Feature Importance for Predicting Marijuana Use (MRJFLAG)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Create dictionary for mapping
readable_feature_names = {
    'IRSEX': 'Gender',
    'NEWRACE2': 'Race/Ethnicity',
    'INCOME': 'Family Income Level',
    'POVERTY3': 'Poverty Status',
    'PDEN10': 'Population Density',
    'COUTYP4': 'Metro/Non-Metro County',
    'HEALTH2': 'Self-Reported Health',
    'EDUSCHLGO': 'Currently in School',
    'EDUSKPCOM': 'School Absence (Sick Days)',
    'EDUSCHGRD2': 'Grade Level',
    'IMOTHER': 'Mother in Household',
    'IFATHER': 'Father in Household',
    'PARCHKHW': 'Parent Checks Homework',
    'PARHLPHW': 'Parent Helps with Homework',
    'PRCHORE2': 'Parent Assigns Chores',
    'PRLMTTV2': 'Parent Limits TV Time',
    'PARLMTSN': 'Parent Limits Screen Use',
    'PRGDJOB2': 'Parent Encourages Job',
    'PRPROUD2': 'Parent is Proud',
    'ARGUPAR': 'Argues with Parents',
    'ALCFLAG': 'Ever Used Alcohol',
    'TOBFLAG': 'Ever Used Tobacco',
    'ALCYDAYS': 'Alcohol Use Days (Year)',
    'CIGMDAYS': 'Cigarette Use Days (Month)',
    'MRJMDAYS': 'Marijuana Use Days (Month)'
}

# Apply mapping to a list of feature names (e.g., for plotting)
original_features = X.columns  # Or replace with your feature list
readable_features = [readable_feature_names.get(feat, feat) for feat in original_features]



# add more features
features = all_features + ['ALCFLAG', 'TOBFLAG', 'ALCYDAYS', 'CIGMDAYS', 'MRJMDAYS']
df_clean = df.copy()

# Replace 991/993 with NaN in relevant columns
df_clean[features + ['IRMJAGE']] = df_clean[features + ['IRMJAGE']].replace([991, 993], np.nan)

# Drop NA
df_clean = df_clean.dropna(subset=features + ['IRMJAGE'])

X_3 = df_clean[features]
y_3 = df_clean['IRMJAGE']

# Split the Data between test and train
X_train_3, X_test_3, y_train_3, y_test_3 =  train_test_split(X_3
                                                 ,y_3
                                                 ,train_size = 0.7, test_size = 0.3
                                                 ,random_state = 1)


# fit Random Forests model
drug_bagging = RandomForestRegressor(max_features=X_train_3.shape[1],random_state = 1)
drug_bagging.fit(X_train_3,y_train_3)

print("Number of trees:", drug_bagging.n_estimators)
print("Number of features tried at each split:",drug_bagging.max_features)
print("Training score: {:.2f}%".format(drug_bagging.score(X_train_3,y_train_3)*100))
```
